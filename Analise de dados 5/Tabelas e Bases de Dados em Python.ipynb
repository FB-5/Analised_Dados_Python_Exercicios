```python
#percorrer todos os arquivos da pasta base de dados (Pasta Vendas)
import os
import pandas as pd
```

```python
lista_arquivo = os.listdir("/content/drive/MyDrive/Analise de dados 2/Vendas")
print(lista_arquivo)
```

```python
tabela_total = pd.DataFrame()
# impostar base de dados de vendas
for arquivo in lista_arquivo:
  # se tem "vendas" no nome do arquivo, então              tab -   print()
  if "Vendas" in arquivo:
    # importar arquivo
    tabela = pd.read_csv(f"/content/drive/MyDrive/Analise de dados 2/Vendas/{arquivo}")
    #.append é estar adicionando uma tebela na outra
    tabela_total = tabela_total.append(tabela)
# tratar e compilar a base de dados
display(tabela_total)
```
